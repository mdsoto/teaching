## 1.4 Image formats

There are many digital formats used to store and transfer images. They are different in terms of compression, quality, color depth and other aspects. Here is a description of the main differences between the main formats (Chat-GPT):

</br>

| Format | Compression | Quality | Color Depth | Transparency | Usage |
| --- |  --- | --- | --- | --- | --- |
| **TIFF** (Tagged Image File Format) | TIFF typically uses lossless compression, meaning there is no loss of image quality when compressed. This results in larger file sizes compared to other formats. | TIFF is known for preserving the original image quality, making it an ideal choice for high-quality images and lossless data retention. | It can support a wide range of color depths, including the CMYK format used in printing. | It can store transparency information and layers. | It is widely used in the printing and professional photography industry due to its high quality and data preservation capabilities. |
| **PNG** (Portable Network Graphics) |  PNG uses lossless compression, meaning there is no loss of image quality, but it tends to be more file size-efficient compared to TIFF. | It also preserves the original image quality and is suitable for images with transparency. | It supports color depths of 8, 24, and 32 bits per pixel. | It efficiently represents transparency, making it ideal for web graphics and elements with transparent backgrounds. | It is commonly used in web design and digital graphics due to its quality and lossless compression capabilities. |
| **JPG** (Joint Photographic Experts Group) | JPG uses lossy compression, meaning there can be a loss of quality when an image is compressed. The amount of compression can be adjusted to balance quality and file size. | While some quality is lost during compression, JPG is suitable for photographs and images with smooth color transitions. It is less suitable for images with high-contrast areas and sharp edges, as it can introduce compression artifacts. | It is generally used with 8 bits per pixel color depth. | JPG does not support transparency. | JPG is widely used in digital photography and on the web due to its adjustable compression, allowing a balance between quality and file size. |
| **GIF** (Graphics Interchange Format) |  GIF uses lossless compression for images, but it is more efficient for images with limited color palettes. For animations, it uses lossy compression. | GIF is suitable for images with limited colors and simple graphics. For animations, it can be limited in terms of quality due to its reduced color palette. | GIF supports color depths of up to 8 bits per pixel for still images. | GIF supports transparency, allowing one color to be transparent. | GIF is commonly used for simple web graphics, animations, and images with transparency, such as icons and memes. It is not suitable for high-quality photographs. |
| **BMP** (Bitmap Image File) |  BMP uses lossless compression or no compression, resulting in potentially large file sizes. | It maintains the original image quality | It supports color depths of 1, 4, 8, 16, 24, or 32 bits per pixel. | BMP does not support transparency. | BMP is often used for simple graphics like icons and logos in Windows systems. Due to its lack of compression, BMP files tend to be larger than other formats.|

</br>

Apart from the basic information, image files usually (not allways) have very detailed information about their origin and history. This information, if is available, is in the image itself, under a data type call EXIF (exchangeable image file). This information is known as **metadata** and allows us to know the history of modifications that the image has undergone, if it has had any. The procedure to access the metadata is below and comes from:

https://faun.pub/making-an-image-metadata-extractor-in-python-40d54f9e9f40

```python
# Extract EXIF or metadata

exifdata = mom_baby_original.getexif()
exifdata
```

```python
# Iterating over all EXIF data fields

for tag_id in exifdata:
    
    # get the tag name and data
    tag = TAGS.get(tag_id, tag_id)
    data = exifdata.get(tag_id)
    
    # decode bytes 
    if isinstance(data, bytes):
        data = data.decode()
    
    print(f"{tag:25}: {data}")
```

## 1.5 Mask and transparency

In the digital photography era, photographic masks are more commonly created and applied digitally using image editing software like Adobe Photoshop. Digital masks are essentially grayscale images where white areas indicate where light is allowed to pass through, and black areas block light. Shades of gray represent partial light transmission (Chat-GPT). Masks can be saved in any format however, the **transparency** of the white areas of the masks is only achieved with the **PNG** format (from Photoshop). Let's compare two masks:

```python
# %matplotlib qt
%matplotlib inline
```

```python
# Loading jpg and png masks of the sky

sky_mask_jpg = Image.open("Input/sky_mask.jpg")

plt.figure(figsize=(11, 8))

plt.subplot(121)
plt.imshow(sky_mask_jpg)
plt.title('JPEG mask')

sky_mask_png = Image.open("Input/sky_mask.png")
plt.subplot(122)
plt.imshow(sky_mask_png)
plt.title('PNG mask with transparency');
```

Activating qt allows you to explore the values of both white areas. In the JPG mask, the value is (255, 255, 255), which is white. In the white area of the PNG mask, the values are (0, 0, 0, 0) which does not correspond to white. The fourth zero corresponds to the alpha channel that controls the transparency. This inconsistency tells us that **there is no information in the white areas of the PNG masks** (from Photoshop).

```python
# Let´s put the masks over the original image

plt.figure(figsize=(11, 8))

plt.subplot(121)
plt.imshow(mom_baby)
plt.imshow(sky_mask_jpg)
plt.title('Image  + JPEG mask')

plt.subplot(122)
plt.imshow(mom_baby)
plt.imshow(sky_mask_png)
plt.title('Image  + PNG mask');
```

<h2> 3.3 Counting pixels </h2>

The Pillow's function `getcolors` reports in a list the frequency of each pixel value in the channel

```python
# Counting pixels on each channel

red_getcolors_list = red.getcolors()
print(red_getcolors_list)

# To simplify the operation it is better to convert the list to an array

red_getcolors_array = np.array(red_getcolors_list)
print(red_getcolors_array)
```

```python
# Creating separated variables for counting and pixel value

red_pixel_count = red_getcolors_array[:,0]
red_pixel_value = red_getcolors_array[:,1]

print(red_pixel_count, '\n')
print(red_pixel_value)
```

```python
# Total amount of pixels

print('Total amount of pixels:', sum(red_pixel_count), '\n')

# Not all pixels values are present in the red channel

print('Different values of pixels:', sum(red_pixel_value))
```

```python
# Adding all white pixels in the red channel

# This operation is a mask or filter in which the value 235 is a cutoff

white_in_red = red_pixel_count[red_pixel_value > 235]    
print('Counting ofthe white pixels:\n', '\n', white_in_red,'\n')

red_mineral_pixels = np.sum(white_in_red)
print('Total of white pixels in the red channel: ', red_mineral_pixels)
```

```python
# Percentage of red mineral, carbonate

total_pixels = 716_000
print('red mineral (%) = ', 100*(red_mineral_pixels/total_pixels))
```
