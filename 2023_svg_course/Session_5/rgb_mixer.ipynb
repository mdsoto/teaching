```python
# Libraries

from PIL import Image
import ipywidgets as widgets
```

```python
# Function for potting

RGB = {'R': 0, 'G': 0, 'B': 0}

def my_color_scale(R, G, B, width, height):

    my_color = (int(R), int(G), int(B))
    monocolor = Image.new('RGB', (int(width), int(height)), my_color)
    
    RGB['R'] = R
    RGB['G'] = G
    RGB['B'] = B
    
    return monocolor
```

```python
# Function for RGB to Hexadecimal convertion

def rgb2hex(r,g,b):
    return "#{:02x}{:02x}{:02x}".format(r,g,b)
```

```python
# RGB mixer

width, height = 1000, 500

widgets.interact(my_color_scale, R=widgets.FloatSlider(value=127, min=0, max=255, step=1),
                                 G=widgets.FloatSlider(value=127, min=0, max=255, step=1),
                                 B=widgets.FloatSlider(value=127, min=0, max=255, step=1),
                                 width=widgets.fixed(width),
                                 height=widgets.fixed(height));
```

```python
# Hexadecimal color scale

print('RGB values:', int(RGB['R']), int(RGB['G']), int(RGB['B']))
print('Hexadecimal color scale:', rgb2hex( int(RGB['R']), int(RGB['G']), int(RGB['B'])))
```
