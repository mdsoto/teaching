<h2> 2.2 Central tendency</h2>

Central tendency focuses on the identification of the central or more representative value in a data set or variable.

```python
# The main parameters associated with the central tendency are:

print('mode           =', stats.mode(zn))    # function of the scipy library, there is no mode in np
print('median         =', round(np.median(zn), 3))
print('mean (average) =', round(np.mean(zn), 3))
```

The formula of mean and others are in the [annex](#annex).

```python
# Median is the central value of the sorted data, let`s define a function to figure out the median

def my_median(x):

    '''Simple function to find the media in a list'''
    
    y = np.sort(x)
    center = len(x)//2 # Whole part
    
    if len(x)%2 == 0: # Remainder
        print('Even list median:', (y[center] + y[center - 1])/2)
    else:
        print('Odd list median:', (y[center]))
```

```python
lista = [1, 2, 3, 4, 5, 6]

# lista = zn

my_median(lista)
```

Once you know how to get the previous statistical parameters you can incorporate them on your plots

<h1> 3 Normal distribution </h1>

To advance further, we have to go back and review some aspects of the most important concept in statistics, the normal or Gaussian distribution. The normal distribution is a function that helps us to model a lot of natural events. Some properties of the normal distribution are:

* It has a characteristic bell-shape
* Symmetrical around a central value (mean), that is at the highest point of the curve
* Mean, median, and mode are the same
* Most of the values a around the mean, values far from the mean are less likely to occur
* The area under the curve is one
* Areas of sectors away from the mean, by one, two, or three standard deviations, are well established
* Mean is zero, standard deviation is one

<h2> 3.1 Box plot and outliers</h2>

The box plot is another popular statistical plot that allows the identification of outliers or anomalies. Interquartile Range (IQR) is defined as p75 - p25. Values that fall beyond its upper (Q3 + 1.5 IQR) and lower limit (Q1 - 1.5 IQR) are considered outliers or anomalies. The following image shows how this plot is related to the normal distribution centered in 0 and with a standard deviation of one:

<img src="NB_images\box_psub_sub_dfcorecore.png" style="width:750px">

Image from: https://towardsdatascience.com/understanding-boxplots-5e2sub_sub_dfcorecore7bcbd51

```python
# Importing libraries for this part

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the Pulacayo data

zn = np.genfromtxt('Input/zn.dat')
```

```python
# Box plot and histogram for a regular (non-normalized) variable

iqr = np.percentile(zn, 75) - np.percentile(zn, 25)  # Inter quartil range

plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 1)

plt.boxplot(zn, showmeans=True, meanline=True, vert=False)

# This plot is made automatically, apart from vertical no other parameters are needed

plt.title("Box plot and histogram a regular variable")

plt.xlim(-10, 40)

plt.subplot(2, 1, 2)
plt.hist(zn, bins=20, density=1)

# Let's put the parameters in the histogram that match those in the box plot

plt.axvline(x=np.percentile(zn, 75) + 1.5*iqr, ls=":", color='black', label='p75 + 1.5iqr')
plt.axvline(x=np.percentile(zn, 75), ls="--", color='black', label='p75')
plt.axvline(x=np.median(zn), ls="-", color='orange', label='median')
plt.axvline(x=np.mean(zn), ls="-", color='red', label='mean')
plt.axvline(x=np.percentile(zn, 25), ls="--", color='orange', label='p25')
plt.axvline(x=np.min(zn), ls=":", color='orange', label='min')
plt.axvline(x=np.percentile(zn, 25) - 1.5*iqr, ls=":", color='grey', label='p25 -1.5iqr')
plt.xlim(-10, 40)
plt.xlabel('% Zn')
plt.ylabel('Probability')
plt.legend()

plt.show()

print('min                =', np.min(zn))
print('max                =', np.max(zn))
print('ipr                =', iqr)
print('mean (average)     =', np.mean(zn))
print('median             =', np.median(zn))
print('standard deviation =', np.std(zn))


# As (Q1 - 1.5IQR) < min so the lower limit of the box plot is the minimum
# We have a point over the upper limit (Q3 + 1.5IQR) of the box plot so this point can be considered as an outlier.
```

<h2> 3.2 Normalization</h2>

Normalization is transforming any distribution or variable to make it appear like a normal distribution (mean of 0 and standard deviation of 1). This transformation allows us to make comparisons of the variable with respect to the normal distributions and use its properties.

<a  id="zscore"></a>

<h3> 3.2.1 Z-score normalization</h3>

The Z-score transformation or normalization is the most important normalization in statistics because is linked with the normal distribution (it is in the formula of the normal distribution). The formula of this normalization is:

$$Z = \frac{x - \mu}{\sigma}$$

The mean of Z will be zero and the standard deviation will be 1. This formula and others are in the [annex](#annex).

Where x is the original variable, &mu; is the mean (average) of x, and &sigma; is the standard deviation of x. Let's apply this transformation to the zinc in the Pulacayo mine data to see the differences in the parameters and plots.

```python
# Box plot and histogram for a z-score normalized variable

z_zn = (zn - np.mean(zn))/ np.std(zn)

# z_zn is the new variable

iqr = np.percentile(z_zn, 75) - np.percentile(z_zn, 25)

plt.figure(figsize=(10,10))

plt.subplot(2,1,1)
plt.boxplot(z_zn, showmeans=True, meanline=True, vert=False)
plt.title("Box plot and histogram for a z-score normalized variable")
plt.xlim(-3,3)

plt.subplot(2,1,2)
plt.hist(z_zn, bins = 59, density = 1)
plt.axvline(x = np.percentile(z_zn, 75) + 1.5*iqr, ls = ":", color='black',label='p75 + 1.5iqr')
plt.axvline(x = np.percentile(z_zn, 75), ls = "--", color='black',label='p75')
plt.axvline(x = np.median(z_zn), ls = "-", color='orange',label='median')
plt.axvline(x = np.mean(z_zn), ls = "-", color='red',label='mean')
plt.axvline(x = np.min(z_zn), ls = ":", color='orange',label='min')
plt.axvline(x = np.percentile(z_zn, 25) - 1.5*iqr, ls = ":", color='grey',label='p25 - 1.5iqr')
plt.xlim(-3,3)
plt.xlabel ('% z_zn')
plt.ylabel ('Probability')
plt.legend()

plt.show()

print('min                =', np.min(z_zn))
print('max                =', np.max(z_zn))
print('mean (average)     =', np.mean(z_zn))
print('median             =', np.median(z_zn))
print('standard deviation =', np.std(z_zn))
```

<a  id="unity"></a>

<h3> 3.2.2 Unity-base normalization</h3>

Also called min-max normalization is a common normalization in well-log analysis. It is defined by the following formula:

$$U = \frac{x - xmin}{xmax - xmin}$$

The values, mean and standard deviation of U will be between 0 and 1. This formula  and others are in the [annex](#annex).

Let's apply this transformation to the zinc in the Pulacayo mine data to see the differences in the box and histogram.

```python
# Box plot and histogram for a unity-base normalized variable

u_zn = (zn - np.min(zn)) / (np.max(zn) - np.min(zn))

# u_zn is the new variable

iqr = np.percentile(u_zn, 75) - np.percentile(u_zn, 25)

plt.figure(figsize=(10, 10))

plt.subplot(2, 1, 1)
plt.boxplot(u_zn, showmeans=True, meanline=True, vert=False)
plt.title("Box plot and histogram for a unity-base normalized variable")
plt.xlim(-0.5, 1.5)

plt.subplot(2, 1, 2)
plt.hist(u_zn, bins=59, density=1)
plt.axvline(x=np.percentile(u_zn, 75) + 1.5*iqr, ls=":", color='black', label='p75 + 1.5iqr')
plt.axvline(x=np.percentile(u_zn, 75), ls="--", color='black', label='p75')
plt.axvline(x=np.median(u_zn), ls="-", color='orange', label='median')
plt.axvline(x=np.mean(u_zn), ls="-", color='red', label='mean')
plt.axvline(x=np.min(u_zn), ls=":", color='orange', label='min')
plt.axvline(x=np.percentile(u_zn, 25) - 1.5*iqr,
            ls=":", color='grey', label='p25 - 1.5iqr')
plt.xlim(-0.5, 1.5)
plt.xlabel('% u_zn')
plt.ylabel('Probability')
plt.legend()

plt.show()

print('min                =', np.min(u_zn))
print('max                =', np.max(u_zn))
print('mean (average)     =', np.mean(u_zn))
print('median             =', np.median(u_zn))
print('standard deviation =', np.std(u_zn))
```

<h2> 3.3 Other distributions</h2>

Although many natural phenomena exhibit behaviors that conform to the normal distribution, there are many other processes that fit well with other types of distributions. Numpy, like other statistical libraries (stat, scipy, ...) allows you to generate values for many types of distributions.

All Numpy distributions at: https://numpy.org/doc/stable/reference/random/generator.html?highlight=distributions <br>
All Scipy distributions at: https://docs.scipy.org/doc/scipy/reference/stats.html

Apart from the normal, here we present three other distributions, one uniform, one Poisson, and one bimodal. Note that each of them has different arguments. The entire explanation at: https://towardsdatascience.com/the-signal-and-the-noise-d82b8630c3ad

```python
# Here is an easy way to write a big number
sample = 10_000_000    

# Uniformly distribution for a 6-sided dice
x = np.random.choice([1,2,3,4,5,6], size = sample)

plt.figure(figsize=(17, 5))

plt.subplot(1,4,1)
plt.hist(x, bins=6)
plt.title('Uniformly distribution')
plt.xlabel('Uniform x ')
plt.ylabel('Frequency')

# Poisson distribution for the number of received letters
x = np.random.poisson(lam = 4, size = sample)

plt.subplot(1,4,2)
plt.hist(x, bins=10)
plt.title('Poisson distribution')
plt.xlabel('Poisson x')

# Binomial Distribution for 5 heads in 10 coin flips
x = np.random.binomial(10, 0.5, size = sample)

plt.subplot(1,4,3)
plt.hist(x, bins=10)
plt.title('Binomial distribution')
plt.xlabel('Binomial x')

# Lognormal distribution
rng = np.random.default_rng()
mu, sigma = 5 , 1. # mean and standard deviation
x = rng.lognormal(mu, sigma, size=sample)

plt.subplot(1,4,4)
plt.hist(np.log10(x), bins=10)
plt.title('Lognormal distribution')
plt.xlabel('log(x)')

plt.tight_layout()
plt.show()
```

<h2> 4.2 Covariance and correlation coefficient</h2>

The quantitative way to estimate the correlation between two variables is by calculating the covariance and correlation coefficient. Covariance is the equivalent of the variance but for two variables. The formula is:

$$Cov(x,y)=\frac{1}{n}\sum_{i=1}^n (x_i - \mu x)(y_i - \mu y)$$

<br>

Covariance can have negative (x and y grow in an inverse way) or positive values (x and y grow together). An inconvenience of this parameter is that it has strange units (unit of x time unit of y) which make its interpretation difficult.

<img src="NB_images\cov.png" style="width:600px">

Image from: https://www.datasciencecentral.com/profiles/blogs/preprocessing-for-deep-learning-from-covariance-matrix-to-image

A better parameter for estimating the relation between two variables is the correlation coefficient, which is a unitless parameter, from 0 (no correlation) to 1 (maximum positive correlation or x and y grow together) or -1 (maximum negative correlation or x and y grow in an inverse way). The formula is:

<br>

$$\rho (x,y) = \frac{Cov(x,y)}{\sigma x \sigma y}$$

<br>

<img src="NB_images\correl.png" style="width:600px">

<br>

Image form: https://en.wikipedia.org/wiki/Pearson_correlation_coefficient

<b>Both parameters are valid only if the variables have a certain degree of linearity (upper two rows of the image above).</b>

In both formulas; x and y are the original variables, &mu;x is the mean of x, &mu;y is the mean of y, 
&sigma;x is the standard deviation of x, and &sigma;y is the standard deviation of y. These formulas and others are in the [annex](#annex).

```python
# For covariance and correlation is worth creating functions

def my_cov(x,y):
    n = len(x)
    return sum((x-np.mean(x))*(y-np.mean(y)))/(n)

def my_correl(x,y):
    n = len(x)
    return sum((x-np.mean(x))*(y-np.mean(y)))/((sum((x-np.mean(x))**2))**0.5 * (sum((y-np.mean(y))**2))**0.5)
```

```python
# Let's do some calculations

print('Covariance of gr vs. he_phi      :',my_cov(sub_dfcore.gr,sub_dfcore.he_phi))
print('Correlation of gr vs. he_phi     :',my_correl(sub_dfcore.gr,sub_dfcore.he_phi))
print()

print('Covariance of he_phi vs. air_k   :',my_cov(sub_dfcore.he_phi, sub_dfcore.air_k))
print('Correlation of he_phi vs. air_k  :',my_correl(sub_dfcore.he_phi, sub_dfcore.air_k))
print()

print('Covariance of air_k vs. klink_k  :',my_cov(sub_dfcore.air_k, sub_dfcore.klink_k))
print('Correlation of air_k vs. klink_k :',my_correl(sub_dfcore.air_k, sub_dfcore.klink_k))
```

## Annex of formulas

In most of the statistical formulas, instead of *n* for the entire population of the random variable x, you can use *n-1* for a smaller sample of x.

Range
$$range = x_{max} - x_{min} $$

Mean (average)
$$\mu=\frac{1}{n}\sum_{i=1}^n x_i$$

Absolute deviation, with mean or median (central value of x)
$$D_i = \mid x_i - \mu \mid$$
$$D_i = \mid x_i - median \mid$$

Variance
$$\sigma^2=\frac{1}{n}\sum_{i=1}^n (x_i - \mu)^2$$

Standard deviation
$$\sigma=\sqrt(\sigma^2)$$

Unity-base normalization
$$U = \frac{x - xmin}{xmax - xmin}$$

Z-score normalization
$$Z = \frac{x - \mu}{\sigma}$$
Normal distribution

$$f(x) = \frac{1}{\sigma\sqrt2\pi}e^-\frac{1}{2}Z^2$$

Covariance

$$Cov(x,y)=\frac{1}{n}\sum_{i=1}^n (x_i - \mu x)(y_i - \mu y)$$

Correlation coefficient

$$\rho (x,y) = \frac{Cov(x,y)}{\sigma x \sigma y}$$
