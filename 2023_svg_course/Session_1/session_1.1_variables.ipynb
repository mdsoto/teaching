#### Python operator precedence and association

The order in which operations are executed in Python follows a well-established order or **precedence**:

https://docs.python.org/3/reference/expressions.html

In the example below we can see that `*` has priority over `-`:

```python
5 * 2 - 4 
```

```python
# This is equivalent to the previous operation:

(5 * 2 ) - 4
```

In the case of operations with the same level of precedence, Python will execute the operation on the left first, this is known as **left association:** 

```python
6 * 2 // 3
```

```python
# This is equivalent to the previous operation:

(6 * 2) // 3
```

The power is an exception of the left association:

```python
2 ** 3 ** 2
```

```python
# This is equivalent to the previous operation:

2 ** (3 ** 2)
```

Since parentheses have the highest precedence, we can force the desired order in the operations by using them.

<h2> 1.2 Variables</h2>

In Python, variables point to reserved memory locations that store values. Along the execution of a program, the variables can have any type (strings, numbers, boolean ...), which are verified only during the execution of an operation or function. This behavior, known as **dynamic typing**, gives Python a lot of flexibility when declaring variables. However, in the end, the variables must agree with the operations you want to perform. In the example below the operator **+** is an example of an **overloading operator** which can perform different operations depending on the type of inputs:

```python
# Variables a and b are integers

a = 2
b = 5
```

```python
a + b
```

```python
a*5
```

```python
# Now they are sequences of strings

a = 'This is'
b = ' dynamic typing'
```

```python
a + b
```

```python
a*5
```

#### Variable names

* Variable names must **start with a letter or the underscore character**, not a number.  
* A variable name can only **contain alpha-numeric characters and underscores** (A-z, 0-9, and _ ), not special characters. 
* Variable names are **case-sensitive** (python, Python, and PYTHON are different variables).  

#### Assigning values to variables

The equal sign (=) assigns a value to a variable. More about variables at: https://realpython.com/python-variables/

```python
# Simple calculation of the area of a circle inside the print()

pi = 3.1416
radius = 100
unit = 'square meters'

print('Area of the circle:', pi*(radius**2), unit)
```

```python
# Hypotenuse of a right triangle

a = 30
b = 40
c = (a**2 + b**2)**0.5

unit = 'cm'  # you can put the quotations before or after the text

print("Hypotenuse:", c, unit)
```
