<h2>2.1 Reading text files </h2>

Text files are the most straightforward files used in any computer (regardless of the operative system) to save information in a format familiar to us. Some specific uses of the text files are:

* Simple documents like notes, letters, emails, ...
* Data in columnar shape
* Keepers of parameters or environment variables for programs or operative systems
* Scripts or programs on different programming languages
* Communication between machines, programs, and persons
* Databases

Characters in text files were initially coded for electronic communication (teleprinters) based on different standards that in 1963 became the **ASCII** (American Standard Code for Information Interchange) code. Soon, it was clear that the ASCII code, which only had English characters, some mathematical,  and special symbols (128 characters), was insufficient for other written languages. So the ASCII code started to evolve into different codes until we got to the actual **UTF-8** (Universal Transformation Format) which has 1,112,064 characters, enough for all human written languages and disciplines. The UTF-8 standard is the base on which the World Wide Web works.

More information on character encoding at: https://www.w3schools.com/charsets/

<h3> 2.1.2 Reading multi-column text files with missing values </h3>

To have a file with just a column is very rare. Normally we have files with several numerical columns, sometimes even columns with text. When you load a multi-column file an array is created with the corresponding number of columns, later each one can be defined as independent variables. Important, to operate with these variables, they have to have the same dimensions (same number of rows).

A common situation is to have several missing or lost values, places where the measurements were not possible. These missing values used to be indicated with special numbers such as -999.000, or -999.2500, or just with empty spaces. Numpy recognizes automatically empty spaces as missing values and replaces them with **nan**, not a number.

We are going to load the *rampi.csv* file, which is a five-column file that comes from a study of gold and associated elements (in ppm) in quartz veins in the **Rampi block prospect**, Indonesia (see image below from Google Earth):

<img src="NB_images\rampi_indonesia.png"  style="width:700px">

Data and information on the area in:

https://www.researchgate.net/publication/321018113_Occurrences_and_Characteristics_of_Gold_Mineralization_in_Rampi_Block_Prospect_North_Luwu_Regency_South_Sulawesi_Province_Indonesia/figures?lo=1

```python
# Loading the multi-column file

rampi = np.genfromtxt('Input/rampi.csv', skip_header = 1, delimiter=',')

# The default data type of the np.genfromtxt command is float64 however you can set the loading to other types

print(rampi, '\n')
print(type(rampi))
rampi.shape
```

```python
# Variables assignation

au = rampi[:,0]
ag = rampi[:,1]
cu = rampi[:,2]
zn2 = rampi[:,3]

# We use zn2 in order to avoid the overwriting of the zn from the Bolivian data

pb = rampi[:,4]

print('Silver:\n', ag, '\n')
print(type(ag))
```

```python
# Some basic operations cannot be executed because some arrays have missing values (nan)

print('samples =',len(ag))
print('min =',np.min(ag))
print('max =',np.max(ag))

ag_range = np.max(ag) - np.min(ag)
print('range =', ag_range)
```

```python
# Arrays with nan require special functions that ignore the missing values

print('samples =',len(ag))

print('min =',np.nanmin(ag))
print('max =',np.nanmax(ag))

ag_range = np.nanmax(ag) - np.nanmin(ag)
print('range =', ag_range)
```

<h3> 3.1.1 Point-line plots </h3>

For this demonstration, we are going to use the same data as the previous section, the data from Pulacayo Mine (*zn.dat*), and from the Rampi prospect block (*rampi.csv*). The `import` and loading commands are repeated in each section of the notebook so you can run each section individually.

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the Pulacayo data, single zinc concentration (weight %)

zn = np.genfromtxt('Input/zn.dat')
print('lenth of the zn file: ', len(zn))
```

```python
# A simple point-line plot of the original data. If x is not declared it assumes the position in the file.

plt.plot(zn)

plt.plot(zn, '.')
```

```python
# Let's plot the concentration of zinc against its real location along the vein, each 2 m

# This will be a cross plot (xy plot) with axes labels.

x = range(0, 2*len(zn), 2)    # remember the sintaxis of range is: range(start, end, step)

plt.figure(figsize=(8, 5))

plt.plot(x, zn)
plt.xlabel("Distance (m)")
plt.ylabel("Weight %")
plt.title("Zn along quartz vein")
plt.show()
```

```python
# Let's show the mean value in the previous plot

y_mean = np.mean(zn) # y is a single value

plt.figure(figsize=(8,5))
plt.plot(x, zn)

# axhline put a horizontal line base on y

plt.axhline(y_mean, ls="-", color='red', label="mean")

plt.legend()

plt.xlabel("Distance (m)")
plt.ylabel("Weight %")
plt.title("Zn along quartz vein")
plt.show()
```

```python
# Let's fill between curves

plt.figure(figsize=(8,5))
plt.plot(x, zn)
plt.axhline(y_mean, ls="-", color='black', label="mean")

plt.fill_between(x, zn, y_mean, where=(zn >= y_mean), facecolor='b', interpolate=True, label="over mean")
plt.fill_between(x, zn, y_mean, where=(zn <= y_mean), facecolor='r', interpolate=True, label="under mean")

plt.xlabel("Distance (m)")
plt.ylabel("Weight %")
plt.title("Zn along quartz vein")
plt.legend()
plt.show()
```

```python
# Loading the Rampi data with five elements concentration (ppm), and variables assignation

rampi = np.genfromtxt('Input/rampi.csv', skip_header = 1, delimiter=',')

au = rampi[:,0]
ag = rampi[:,1]
cu = rampi[:,2]
zn2 = rampi[:,3]
pb = rampi[:,4]
```

```python
# Multi point-line plot of the original data, just in the same file order, no x is involved, with grid.

plt.figure(figsize=(8,5))

plt.plot(au, 'y', label="Au")
plt.plot(ag, 'c', label="Ag")
plt.plot(cu, 'r', label="Cu")
plt.plot(zn2,'m', label="Zn")
plt.plot(pb, 'b', label="Pb")

plt.title("Rampi elements")
plt.xlabel("File order")
plt.ylabel("Concentration(ppm)")
plt.legend()
plt.grid(True)
plt.show()
```

<h3> 3.1.3 Bar and pie plots </h3>

Here you have two more types of plots, bar and pie plots. The categories (elements), data (sizes), and colors are the same in both cases.

<h2> 3.2 Multiplots by subplot </h2>

With the Matploplib's **subplot** function it is possible to gather several plots in a composition, arranged based on a rectangular grid of raw and columns. The arguments (2,2,1) in the following command:

`plt.subplot(2,2,1)`
    
means the first plot in an arrange of two rows and two columns of plots, something like this:

<img src="NB_images\multiplot_example.png"  style="width:700px">

```python
# Building and saving a 2 x 3 multiplot with the Rampi data

plt.figure(figsize=(16,10))

plt.subplot(2,3,1)

# Point_line plot
plt.plot(au, 'y', label="Au")
plt.plot(ag, 'c', label="Ag")
plt.plot(cu, 'r', label="Cu")
plt.plot(zn2,'m', label="Zn")
plt.plot(pb, 'b', label="Pb")
plt.title("Elements")
plt.xlabel("File order")
plt.ylabel("ppm")
plt.legend()
plt.grid(True)

plt.subplot(2,3,2)

# Bar plot
elements = ('Au', 'Ag', 'Cu', 'Zn', 'Pb')
sizes = [np.nanmean(au), np.nanmean(ag), np.nanmean(cu), np.nanmean(zn2), np.nanmean(pb)]
x = np.arange(len(elements))   
colors = ['y', 'c', 'r','m','b']
plt.bar(x, sizes, align='center', alpha=0.5, color=colors)
plt.xticks(x, elements)
plt.title('Elements')
plt.ylabel('ppm')
plt.xlabel('Elements')
plt.grid(True)

plt.subplot(2,3,3)

# Cross plot
plt.loglog(au,ag,'o')
plt.title("Au vs. Ag")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Ag (ppm)")
plt.grid(True)

plt.subplot(2,3,4)

# Cross plot
plt.loglog(au,cu,'o')
plt.title("Au vs. Cu")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Cu (ppm)")
plt.ylim(1,200)
plt.grid(True)

plt.subplot(2,3,5)

# Cross plot
plt.loglog(au,zn2,'o')
plt.title("Au vs. Zn")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Zn (ppm)")
plt.ylim(1,200)
plt.grid(True)

plt.subplot(2,3,6)

# Cross plot
plt.loglog(au,pb,'o')
plt.title("Au vs. Pb")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Pb (ppm)")
plt.ylim(1,200)
plt.grid(True)

# Avoid the overlap between labels

plt.tight_layout()

# Saving multiplot to file

plt.savefig("Output/multiplot.pdf")
plt.savefig("Output/multiplot.png")
plt.show()
```

<h2> 3.3 Multiplots by loop </h2>

Multiplots can be built by a loop, making the code shorter and more efficient. Let's see an example with the Rampi data: 

```python
# Elements, colors, and position in the vein, according to the rampi array

elements = ('Au', 'Ag', 'Cu', 'Zn', 'Pb')
colors = ['y', 'c', 'r', 'm', 'b']
x = range(0, 3*rampi.shape[0], 3)

# Loop according to the number of columns in the rampi array

for i in range(rampi.shape[1]):
    
    plt.figure(figsize=(16,4))
    
    # Line plot of the element
    plt.subplot(1,2,1)
    plt.plot(x, rampi[:,i], color=colors[i], alpha=0.5)
    plt.xlim(0, 65)
    plt.xlabel('Position in the vein(m)')
    plt.ylabel('Concentration(ppm)')
    plt.title('Plot of '+elements[i])
    plt.grid()
    
    # Histogram of the element
    plt.subplot(1,2,2)
    plt.hist(rampi[:,i], color=colors[i], alpha=0.5)
    plt.ylabel('Frequency')
    plt.xlabel('Concentration(ppm)')
    plt.title('Histogram of '+elements[i])
    plt.grid()
    plt.savefig("Output/plot_"+str(elements[i])+".png")  
     
# plt.tight_layout()
# plt.savefig("Output/multiplot_loop.png")      
plt.show()        
```

In summary, we have:

| Multiplot by | Number of plots | Number of effective lines of code |
| --- | --- |--- |
| Pure subplot, part 3.2 of this NB| 6 | 52 |
| For loop | 10 | 20 |

The way forward is clear. Whenever possible, try to use for loops to build your multiplots, especially if you are repeating the same set of plots.

```python
# Importing libraries

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from matplotlib.widgets import Cursor
```

```python
# Sine plot with an interactive slider

# Plotting function

def plot_func(freq):
    plt.figure(figsize=(11,5))
    x = np.arange(0, 2*np.pi, 0.01) # 2*np.pi is 360º
    y = np.sin(x * freq)
    plt.plot(x, y)
    plt.xlabel('Radians')
    plt.ylabel('Amplitude')
    plt.grid()
    
# The widget receives the plotting function

widgets.interact(plot_func, freq = widgets.FloatSlider(value=1, min=1, max=20, step=0.1))
# widgets.interact(plot_func, freq = widgets.IntSlider(value=1, min=1, max=20, step=1))
plt.show()
```

## 4.2 Procedural vs. object-oriented programming

Before covering the last topic is worth seeing the difference between building a multiplot using the procedural and the object-oriented programming (OOP) paradigms. Here we are going to do the same scatter plot, with the same data, with both paradigms of coding:

```python
# Plot built with the procedural paradigm

plt.figure(figsize=(8,3))
plt.scatter(au, ag)    # The following three lines are equivalent to plt.loglog(au,ag,'.')
plt.yscale('log')
plt.xscale('log')
plt.title("Au vs. Ag")
plt.xlabel("Log Au (ppm)")
plt.ylabel("Log Ag (ppm)")
plt.grid(True)
plt.show()
```

```python
# Plot built with the OOP paradigm

fig, ax = plt.subplots(figsize=(8, 3))
ax.scatter(au,ag)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_title("Au vs. Ag")
ax.set_xlabel("Log Au (ppm)")
ax.set_ylabel("Log Ag (ppm)")
ax.grid(True)
plt.show()
```

In code length both styles are equivalent. In the availability of examples and power, the **oop paradigm** (more modern) **is the winner** and is where we have to aim.
