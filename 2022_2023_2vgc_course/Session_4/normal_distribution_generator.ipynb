```python
# Importing the libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# Random normal distribution generator

mu, sigma, bins_plot, samples = 0, 1, 1000, 10000 # Vary these numbers; mu (mean), sigma (standard deviation), bins_plot

var = np.random.normal(mu, sigma, samples)

# np.random.normal creates a normal random distribution with mean of mu and standard deviation of sigma
```

```python
var
```

```python
# Point-line plot

x = (range(0, samples, 1))

y1 = mu + sigma
y2 = mu
y3 = mu - sigma

plt.figure(figsize=(13, 8))

plt.plot(x, var)

plt.axhline(y=y1, ls="--", color='black', label="mean + std")
plt.axhline(y=y2, ls="-", color='red', label="mean")
plt.axhline(y=y3, ls=":", color='black', label="mean - std")

plt.title("Normal variable")
plt.xlabel("samples")
plt.ylabel("value")
plt.legend()
plt.show()
```

```python
# Histogram (PDF)

plt.figure(figsize=(13, 8))

# Here the histogram returns variables
count, bins, ignored = plt.hist(var, bins_plot, density=True)

# count and bins are arrays of bins_plots dimentions

plt.axvline(x=y1, ls="--", color='black', label='mean + std')
plt.axvline(x=y2, ls="-", color='red', label='mean')
plt.axvline(x=y3, ls=":", color='black', label='mean - std')

# # Calculation of the theoretical random normal distribution with the parameters provided by the histogram

normal_dist = ((1/(sigma * np.sqrt(2 * np.pi)) *
                np.exp(- (bins - mu)**2 / (2 * sigma**2))))

# Plot of the theoretical random normal distribution

plt.plot(bins, normal_dist, linewidth=2, color='r')

plt.xlabel("value")
plt.ylabel("probability")
plt.show()
```

```python
# CDF

plt.figure(figsize=(13, 8))

plt.hist(var, bins_plot, density=True, cumulative=1)

plt.axvline(x=y1, ls="--", color='black', label='mean + std')
plt.axvline(x=y2, ls="-", color='red', label='mean')
plt.axvline(x=y3, ls=":", color='black', label='mean - std')

plt.xlabel("value")
plt.ylabel("probability")
plt.show()
```

```python
# Bar plot

iqr = np.percentile(var, 75) - np.percentile(var, 25)

plt.figure(figsize=(13, 16))

plt.subplot(2, 1, 1)
plt.boxplot(var, vert=False)
plt.title("Box plot and histogram a regular variable")

# Histogram with parameters

plt.subplot(2, 1, 2)
plt.hist(var, bins_plot, density=True)

plt.axvline(x=np.percentile(var, 75) + 1.5*iqr, ls=":",
            color='black', label='p75 + 1.5iqr')
plt.axvline(x=np.percentile(var, 75), ls="--", color='black', label='p75')
plt.axvline(x=np.median(var), ls="-", color='orange', label='median')
plt.axvline(x=y2, ls="-", color='red', label='mean')
plt.axvline(x=np.percentile(var, 25), ls="--", color='orange', label='p25')
plt.axvline(x=np.percentile(var, 25) - 1.5*iqr, ls=":",
            color='orange', label='p25 - 1.5iqr')
plt.xlabel('value')
plt.ylabel('probability')
plt.legend()
plt.show()

# Parameters

print('Main parameters\n')
print('min = ', np.min(var))
print('max = ', np.max(var))
print('range = ', np.max(var) - np.min(var))
print('median = ', np.median(var))
print()
print('mean (input) = ', mu)
print('mean (output) = ', np.mean(var))
print('standard deviation (input) = ', sigma)
print('standard deviation (output) = ', np.std(var))
print()
print('kurtosis = ', stats.kurtosis(var))
print('skewness = ', stats.skew(var))
```
